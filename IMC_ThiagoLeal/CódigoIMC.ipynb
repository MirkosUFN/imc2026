```python
import pandas as pd

# Média de idade

caminho = '/content/drive/MyDrive/20 Pessoas/planilha_pessoas.xlsx'

df = pd.read_excel(caminho)

print(df.head())

media_idade = df["Idade"].mean()

print("A média de idade das 20 pessoas é:", media_idade)
```

```python
# Menor peso

df = pd.read_excel(caminho)

menor_peso = df["Peso (kg)"].min()

print("O menor peso da planilha é:", menor_peso, "kg")
```

```python
# Maior altura

df = pd.read_excel(caminho)

maior_altura = df["Altura (m)"].max()

print("A maior altura da planilha é:", maior_altura, "m")
```

```python
# Criar o IMC

df = pd.read_excel(caminho)

df["IMC"] = df["Peso (kg)"] / (df["Altura (m)"] ** 2)

print(df)

df.to_excel(caminho, index=False)
```

```python
# Grupo IMC dominante

df = pd.read_excel(caminho)

df["IMC"] = df["Peso (kg)"] / (df["Altura (m)"] ** 2)

def classificar_imc(imc):
    if imc < 18.5:
        return "Baixo peso"
    elif imc < 25:
        return "Peso normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidade"

df["Grupo IMC"] = df["IMC"].apply(classificar_imc)

grupo_dominante = df["Grupo IMC"].mode()[0]

print("O grupo de IMC dominante é:", grupo_dominante)
```

```python
# Toda a planilha

df = pd.read_excel(caminho)

df["IMC"] = df["Peso (kg)"] / (df["Altura (m)"] ** 2)

def classificar_imc(imc):
    if imc < 18.5:
        return "Baixo peso"
    elif imc < 25:
        return "Peso normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidade"

df["Grupo IMC"] = df["IMC"].apply(classificar_imc)

media_idade = df["Idade"].mean()

menor_peso = df["Peso (kg)"].min()

maior_altura = df["Altura (m)"].max()

grupo_dominante = df["Grupo IMC"].mode()[0]

print("Média de idade:", media_idade)
print("Menor peso:", menor_peso, "kg")
print("Maior altura:", maior_altura, "m")
print("Grupo IMC dominante:", grupo_dominante)

print("\nPlanilha completa:")
print(df)
```
