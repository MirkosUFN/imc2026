```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
import pandas as pd

creds, _ = default()
gc = gspread.authorize(creds)

# Abrir a planilha pelo ID
SHEET_ID = "1tyIhWt1w2l9cBUqVY70R8th-WVgK9jsq-jbaWdc9Wb8"
planilha = gc.open_by_key(SHEET_ID)
aba = planilha.sheet1

dados = aba.get_all_records()
df = pd.DataFrame(dados)

print(df)
print()


#1 Idade Média das pessoas
media_idade = df["Idade"].mean()
print(f"Idade Média: {media_idade:.1f} anos")

#2 Menor peso
menor_peso = df["Peso (kg)"].min()
print(f"Menor peso: {menor_peso} kg")

#3 Maior altura
maior_altura = df["Altura (m)"].max()
print(f"Maior altura: {maior_altura} m")

#4 Grupo de IMC dominante
def classificar_imc(imc):
    if imc < 18.5:
        return "Abaixo do peso"
    elif imc < 25:
        return "Peso normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidade"

df["Grupo IMC"] = df["IMC"].apply(classificar_imc)
grupo_dominante = df["Grupo IMC"].value_counts().idxmax()
contagem = df["Grupo IMC"].value_counts().max()
print(f"Grupo IMC que domina: {grupo_dominante}")
print()
print(df["Grupo IMC"].value_counts())
```
