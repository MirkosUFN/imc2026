```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Ciência de Dados/Tabela de Dados Pessoais e Opções - Tabela de Dados Pessoais e Opções.csv')

print(df.columns.tolist())

df['PESO (kg)'] = df['PESO (kg)'].str.replace(',', '.').astype(float)
df['ALTURA (m)'] = df['ALTURA (m)'].str.replace(',', '.').astype(float)

media_idade = df['IDADE (anos)'].mean()

print(f"A média de idade do grupo é de {media_idade:.1f} anos.")

menor_peso = df['PESO (kg)'].min()

print(f"O menor peso do grupo é de {menor_peso} kg.")

maior_altura = df['ALTURA (m)'].max()

print(f"A maior altura do grupo é de {maior_altura}m.")

df['IMC'] = df['PESO (kg)'] / (df['ALTURA (m)'] ** 2)
df['IMC'] = df['IMC'].round(2)

faixas = [0, 18.5, 24.9, 29.9, 100]
nomes_grupos = ['Abaixo do peso', 'Peso normal', 'Sobrepeso', 'Obesidade']
df['GRUPO_IMC'] = pd.cut(df['IMC'], bins=faixas, labels=nomes_grupos)

grupo_dominante = df['GRUPO_IMC'].value_counts().idxmax()

print("Colunas separadas com sucesso!")
print(f"O grupo IMC dominante nas 20 pessoas é: {grupo_dominante}")
```
