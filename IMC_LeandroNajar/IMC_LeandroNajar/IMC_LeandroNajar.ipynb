```python
#ABRIR ARQUIVOS DE DADOS E LER
import pandas as pd
caminho = '/content/drive/MyDrive/Ciencia de Dados/pessoasMedidas.xlsx'
arquivo = pd.read_excel(caminho)
arquivo.info()
```

```python
arquivo=arquivo.fillna(0)
arquivo.describe()
```

```python
arquivo=arquivo.fillna(0)
```

```python
arquivo[['Nome', 'Idade', 'Peso', 'Altura']]
```

```python
media_idade = arquivo['Idade'].mean()
print(f"Média de idade: {media_idade:.1f} anos")
```

```python
menor_peso = arquivo['Peso'].min()
print(f"Menor peso: {menor_peso} kg")
```

```python
maior_altura = arquivo['Altura'].max()
print(f"Maior altura: {maior_altura} m")
```

```python
arquivo['IMC'] = arquivo['Peso'] / (arquivo['Altura'] ** 2)
arquivo['IMC'] = arquivo['IMC'].round(2)

print("\n", arquivo[['Nome', 'Peso', 'Altura', 'IMC']].head())

#salvar
arquivo.to_excel(caminho, index=False)
```

```python
def categorizar_imc(imc):
    if imc < 18.5:
        return 'Abaixo do peso'
    elif imc <= 24.9:
        return 'Peso normal'
    elif imc <= 29.9:
        return 'Sobrepeso'
    else:
        return 'Obeso'

arquivo['Categoria IMC'] = arquivo['IMC'].apply(categorizar_imc)

print("\n", arquivo[['Nome', 'Peso', 'Altura', 'IMC', 'Categoria IMC']].head())

#Salvar
arquivo.to_excel(caminho, index=False)
```

```python
def contar_vogais(nome):
    return sum(1 for letra in nome.lower() if letra in 'aeiou')

arquivo['Vogais no Nome'] = arquivo['Nome'].apply(contar_vogais)

arquivo.to_excel(caminho, index=False)
```

```python
arquivo['Relação Idade/IMC'] = (arquivo['Idade'] / arquivo['IMC']).round(2)

print("\n", arquivo[['Nome', 'IMC', 'Categoria IMC', 'Vogais no Nome', 'Relação Idade/IMC']])

arquivo.to_excel(caminho, index=False)
```
