```python
import pandas as pd

caminho = '/content/drive/MyDrive/Ciência de Dados/pessoa.xlsx'
arquivo = pd.read_excel(caminho)

arquivo.info()

# Qual a média de idade dessas 20 pessoas
media_idade = arquivo['Idade'].mean()
print("Média de idade:", media_idade)

# Qual o menor peso
menor_peso = arquivo['Peso (kg)'].min()
print("Menor peso:", menor_peso)

# Qual a menor altura
maior_altura = arquivo['Altura (m)'].max()
print("Maior altura:", maior_altura)

# Criar a coluna IMC
# peso (kg) / altura (m) ao quadrado
arquivo['IMC'] = arquivo['Peso (kg)'] / (arquivo['Altura (m)'] * arquivo['Altura (m)'])

#print(arquivo)

# Classificação do IMC
def classificar_imc(imc):
    if imc < 18.5:
        return "Abaixo do peso"
    elif imc < 25:
        return "Peso normal"
    elif imc < 30:
        return "Sobrepeso"
    elif imc < 40:
        return "Obesidade"
    else:
        return "Obesidade Grave"

arquivo['Grupo_IMC'] = arquivo['IMC'].apply(classificar_imc)

# Qual o grupo IMC dominande nas 20 pessoas
grupo_dominante = arquivo['Grupo_IMC'].value_counts().idxmax()

print("Grupo de IMC dominante:", grupo_dominante)
```

```python
#Número de vogais em cada nome
def contar_vogais(nome):
    vogais = "aeiouAEIOU"
    contador = 0
    for letra in nome:
        if letra in vogais:
            contador += 1
    return contador

arquivo['Qtd_Vogais'] = arquivo['Nome'].apply(contar_vogais)

print(arquivo)
```

```python
#Relação Idade/IMC
arquivo['Relacao Idade IMC'] = arquivo['Idade'] / arquivo['IMC']

print(arquivo)
```
