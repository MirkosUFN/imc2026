```python
import pandas as pd
arquivo = pd.read_excel('/content/drive/MyDrive/Ciência de dados/Aula3.xlsx')
```

```python
for c in range(0,20):
  pessoa = arquivo.iloc[c]
  imc  = pessoa['PESO'] / (pessoa['ALTURA']*pessoa['ALTURA'])
  if imc > 40:
    arquivo.loc[c, 'IMC'] = 'Obesidade Grau III'
  elif 35 <= imc <= 39.9:
    arquivo.loc[c, 'IMC'] = 'Obesidade Grau II'
  elif 30 <= imc <= 34.9:
    arquivo.loc[c, 'IMC'] = 'Obesidade Grau I'
  elif 25 <= imc <= 29.9:
    arquivo.loc[c, 'IMC'] = 'Sobrepeso'
  elif 18.6 <= imc <= 24.9:
    arquivo.loc[c, 'IMC'] = 'Peso Normal'
  else:
    arquivo.loc[c, 'IMC'] = 'Baixo peso'

arquivo['IMC']
```

```python
arquivo.to_excel('/content/drive/MyDrive/Ciência de dados/filosofos_imc.xlsx')
```
