```python
import pandas as pd

def categoriza(imc):
  if imc < 18.5:
    return 'abaixo do peso'
  if imc >= 18.5 and imc < 24.9:
    return 'peso normal'
  if imc >= 24.9 and imc < 29.9:
    return 'acima do peso'
  if imc >= 29.9 and imc < 34.9:
    return 'obesidade grau I'
  if imc >= 34.9 and imc < 39.9:
    return 'obesidade grau II'
  if imc >= 39.9:
    return 'obesidade grau III'

path = '/content/drive/MyDrive/2026/Ciência de Dados/'
arquivo = pd.read_excel(path+'imc-pandas.xlsx')
#arquivo.info()
print('Media de idade das pessoas',arquivo['idade'].mean())
print('Menor peso:',arquivo['peso'].min())
print('Maior altura:',arquivo['altura'].max())
#vamos criar a lista de valores de imc e colocar em um arquivo xlsx
valores_imc = []
categorias_imc = []
for indice in range(len(arquivo)):
  pessoa = arquivo.iloc[indice]
  imc = pessoa['peso']/((pessoa['altura']*pessoa['altura']))
  valores_imc.append(imc)
  categorias_imc.append(categoriza(imc))
arquivo['IMC']=valores_imc
arquivo['CATEGIMC']=categorias_imc
print('Menor IMC:',arquivo['IMC'].min())

retorno = arquivo['CATEGIMC'].mode()[0]#pega o nome do IMC com mais ocorrências
#print(retorno)
contagem = (arquivo['CATEGIMC']==retorno).sum()#by Bernardo
print(retorno,contagem)
#grava o arquivo com os indices de IMC e suas respectivas categorias
arquivo.to_excel(path+'imc-pandas-imc.xlsx')
```
