```python
import pandas as pd

caminho = '/content/drive/MyDrive/Ciencia de Dados/dados_pessoas.xlsx'
arquivo = pd.read_excel(caminho)


if 'Qtd_Vogais' in arquivo.columns:
    arquivo = arquivo.drop(columns=['Qtd_Vogais'])

# 1. Funções de Estatísticas
def calcular_media_idade(df):
    return df['Idade'].mean()

def calcular_menor_peso(df):
    return df['Peso'].min()

def calcular_maior_altura(df):
    return df['Altura'].max()

media_idade = calcular_media_idade(arquivo)
menor_peso = calcular_menor_peso(arquivo)
maior_altura = calcular_maior_altura(arquivo)

# Função para contar vogais
def contar_vogais(texto):
    vogais = 'aeiouAEIOU'
    return sum(1 for letra in str(texto) if letra in vogais)

# Criando a coluna
if 'Nome' in arquivo.columns:
    arquivo['Vogais'] = arquivo['Nome'].apply(contar_vogais)

print(f'Média de Idade: {media_idade:.2f}')
print(f'Menor Peso: {menor_peso}')
print(f'Maior Altura: {maior_altura}')

# 2. Cálculo do IMC e Categoria Dominante
imcs = []
for i in range(len(arquivo)):
    imc = arquivo['Peso'][i] / (arquivo['Altura'][i] ** 2)
    imcs.append(imc)

arquivo['IMC'] = imcs

# Criando a coluna Idade/IMC Ratio
arquivo['Idade_IMC_Ratio'] = arquivo['Idade'] / arquivo['IMC']

# Salvando as alterações no Excel
arquivo.to_excel('/content/drive/MyDrive/Ciencia de Dados/dados_pessoas.xlsx', index=False)

def verificar_contagem_detalhada(df):
    categorias = []
    for imc in df['IMC']:
        if imc < 18.5: categorias.append('Abaixo do peso')
        elif imc < 25: categorias.append('Normal')
        elif imc < 30: categorias.append('Sobrepeso')
        else: categorias.append('Obeso')

    contagem = pd.Series(categorias).value_counts()
    return contagem

display(verificar_contagem_detalhada(arquivo))
dominante = verificar_contagem_detalhada(arquivo)
print(f'A categoria de IMC dominante é: {dominante.idxmax()}')
display(arquivo.describe())
```
